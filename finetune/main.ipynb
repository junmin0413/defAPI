```python
import os
import torch
import wandb
import evaluate
import numpy as np
from datasets import load_dataset

from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments
)

from peft import LoraConfig
from trl import SFTTrainer
```

### Í≤ĹŽ°ú žĄ§ž†ē

```python
base_model  = "deepseek-ai/deepseek-coder-1.3b-instruct" 

train_path= "/Users/jangjunmin/Documents/code/defAPI_project/Create_data/ft_data.jsonl"
test_path = "/Users/jangjunmin/Documents/code/defAPI_project/Create_data/test_data.jsonl"

new_model = "deekseek-coder-1.3b-instruct-finetuned-gbsw"
```

### Parameter žĄ§ž†ē

```python
lora_r = 32 # ÍįížĚī ŪĀīžąėŽ°Ě ŽćĒ ŽßéžĚÄ žąėž†ē žĚľžĖīŽāė Ž™®ŽćłžĚī Ž≥Ķžě°Ūēīžßź
lora_alpha = 8 # LoRAÍįÄ ž†Āžö©Žź†ŽēĆ ÍįÄž§ĎžĻėžóź žĖľŽßąŽāė žėĀŪĖ•žĚĄ ŽĮłžĻ†žßÄ
lora_dropout = 0.06 # Í≥ľž†ĀŪē© Žį©žßÄ žúĄŪēī ŽďúŽ°≠žēĄžõÉ ŽĻĄžú® žĄ§ž†ē
```

```python
use_4bit = True #4bit žĖĎžěźŪôĒ žā¨žö© žó¨Ž∂Ä
bnb_4bit_compute_dtype = "float16" #dtype žĄ§ž†ē
bnb_4bit_quant_type = "nf4" #žĖĎžěźŪôĒ žú†Ūėē žĄ§ž†ē
use_nested_quant = False #ž§Ďž≤© žĖĎžěźŪôĒ žā¨žö© žó¨Ž∂Ä
```

```python
output_dir = "./results" #Ž™®Žćł žėąžł°Í≤įÍ≥ľ, ž≤īŪĀ¨ŪŹ¨žĚłŪäł ž†Äžě• ŽĒĒŽ†ČŪÜ†Ž¶¨

num_train_epochs = 1 # ŪēôžäĶ epoch žąė
per_device_train_batch_size = 1 # ŪõąŽ†®žö© ŽįįžĻė ŪĀ¨Íłį
per_device_eval_batch_size = 1  # ŪŹČÍįÄžö© ŽįįžĻė ŪĀ¨Íłį
gradient_accumulation_steps = 2  # Í∑łŽěėŽĒĒžĖłŪäłŽ•ľ ŽąĄž†ĀŪē† žóÖŽćįžĚīŪäł žä§ŪÖĚ Ūöüžąė
gradient_checkpointing = True  # Í∑łŽěėŽĒĒžĖłŪäł ž≤īŪĀ¨ŪŹ¨žĚłŪäł ŪôúžĄĪŪôĒ
max_grad_norm = 0.6  # Í∑łŽěėŽĒĒžĖłŪäł ŪĀīŽ¶¨ŪēĎžĚÄ Í∑łŽěėŽĒĒžĖłŪäłžĚė ŪĀ¨ÍłįŽ•ľ ž†úŪēúŪēėžó¨ ŪõąŽ†® ž§Ď žēąž†ēžĄĪžĚĄ ŽÜížěĄ.
learning_rate = 2e-6 # žīąÍłį ŪēôžäĶŽ•† AdamW žėĶŪčįŽßąžĚīž†Ä
weight_decay = 0.001 # bias/LayerNorm ÍįÄž§ĎžĻėŽ•ľ ž†úžôłŪēėÍ≥† Ž™®Žď† Ž†ąžĚīžĖīžóź ž†Āžö©Ūē† Weight decay Íįí
optim = "paged_adamw_32bit"  # žėĶŪčįŽßąžĚīž†Ä žĄ§ž†ē
lr_scheduler_type = "cosine"  # ŪēôžäĶŽ•† žä§žľÄž§ĄŽü¨žĚė žú†Ūėē žĄ§ž†ē, žó¨ÍłįžĄúŽäĒ žĹĒžā¨žĚł žä§žľÄž§ĄŽü¨ žā¨žö©
max_steps = -1 # ŪõąŽ†® žä§ŪÖĚ žąė(num_train_epochs žě¨ž†ēžĚė)
warmup_ratio = 0.03 # (0Ž∂ÄŪĄį learning rateÍĻĆžßÄ) ŪēôžäĶ žīąÍłįžóź ŪēôžäĶŽ•†žĚĄ ž†źžßĄž†ĀžúľŽ°ú ž¶ĚÍįÄžčúŪā§ linear warmup žä§ŪÖĚžĚė Ratio
group_by_length = True  # žčúŪÄÄžä§Ž•ľ ŽŹôžĚľŪēú ÍłłžĚīžĚė ŽįįžĻėŽ°ú Í∑łŽ£ĻŪôĒ, Ž©ĒŽ™®Ž¶¨ ž†ąžēĹ ŽįŹ ŪõąŽ†® žÜćŽŹĄŽ•ľ ŽÜížěĄ
save_steps = 0  # X žóÖŽćįžĚīŪäł Žč®Í≥ĄŽßąŽč§ ž≤īŪĀ¨ŪŹ¨žĚłŪäł ž†Äžě•
logging_steps = 500  # Žß§ X žóÖŽćįžĚīŪäł žä§ŪÖĚ Ž°úÍ∑ł
```

```python
# žĶúŽĆÄ žčúŪÄÄžä§ ÍłłžĚī žĄ§ž†ē
max_seq_length = 1024

# ŽŹôžĚľŪēú žěÖŽ†• žčúŪÄÄžä§žóź žó¨Žü¨ ÍįúžĚė žßßžĚÄ žėąž†úŽ•ľ ŽĄ£žĖī Ūö®žú®žĄĪžĚĄ ŽÜížĚľ žąė žěąžĚĆ
packing = False  

# GPU 0 ž†Ąž≤ī Ž™®Žćł Ž°úŽďú 
device_map = {"": 0}
```

```python
os.environ["WANDB_PROJECT"]="defAPI"
os.environ["WANDB_LOG_MODEL"]="true"
os.environ["WANDB_WATCH"]="false"
```

```python
bnb_config = BitsAndBytesConfig(
  load_in_4bit = use_4bit,
  bnb_4bit_quant_type = bnb_4bit_quant_type,
  bnb_4bit_compute_dtype= getattr(torch, bnb_4bit_compute_dtype),
  bnb_4bit_use_double_quant= use_nested_quant
)
```

```python
dataset = load_dataset("json",
                       data_files={
                         "train": train_path,
                          "test": test_path})
print("="*100)
print(dataset["train"][0])
print("="*100)
print(dataset["test"][0])
```

```python
# ŽćįžĚīŪĄįžÖč ž†ēŽ≥ī ŪôēžĚł
print("="*100)
print(f"ŪõąŽ†® ŽćįžĚīŪĄįžÖč ŪĀ¨Íłį: {len(dataset['train'])} žÉėŪĒĆ")
print("\nŪõąŽ†® ŽćįžĚīŪĄįžÖč ŪäĻžĄĪ:")
print(dataset["train"].features)
print("="*100)
print(f"ŪÖĆžä§Ūäł ŽćįžĚīŪĄįžÖč ŪĀ¨Íłį: {len(dataset['test'])} žÉėŪĒĆ")
print("\nŪÖĆžä§Ūäł ŽćįžĚīŪĄįžÖč ŪäĻžĄĪ:")
print(dataset["test"].features)
print("="*100)
```

```python
model = AutoModelForCausalLM.from_pretrained(
  base_model,
  quantization_config=bnb_config,
  device_map="mps"
)

model.config.use_cache = False
model.config.pretraining_tp = 1
```

```python
tokenizer = AutoTokenizer.from_pretrained(
  base_model,
  trust_remote_code=True
)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
```

```python
def preprocess_fn(example):
    inputs = example["prompt"] + example["completion"]
    tokenized = tokenizer(
        inputs,
        truncation=True,
        max_length=max_seq_length
    )
    
    return tokenized

tokenized_dataset = dataset.map(preprocess_fn, remove_columns=dataset["train"].column_names)
```

```python
train_data = tokenized_dataset["train"].shuffle(seed=42)
test_data = tokenized_dataset["test"].shuffle(seed=42)
```

```python
peft_params = LoraConfig(
  lora_alpha=lora_alpha,
  lora_dropout=lora_dropout,
  r=lora_r,
  bias="none",
  task_type="CAUSAL_LM"
) 
```

```python
training_params = TrainingArguments(
    output_dir=output_dir,
    report_to="wandb",
    num_train_epochs=num_train_epochs,
    per_device_train_batch_size=per_device_train_batch_size,
    gradient_accumulation_steps=gradient_accumulation_steps,
    optim=optim,
    save_steps=save_steps,
    logging_steps=logging_steps,
    learning_rate=learning_rate,
    weight_decay=weight_decay,
    fp16=False,
    bf16=False,
    max_grad_norm=max_grad_norm,
    max_steps=max_steps,
    warmup_ratio=warmup_ratio,
    group_by_length=group_by_length,
    lr_scheduler_type=lr_scheduler_type,
)
```

```python
bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_eval(eval_pred):
  logits, labels = eval_pred

  predictions = np.argmax(logits, axis=-1)
  
  bleu_score = bleu.compute(predictions=predictions, references=labels)
  rouge_score = rouge.compute(predictions=predictions, references=labels)
  accuracy_score = accuracy.compute(predictions=predictions, references=labels)
  f1_score = f1.compute(predictions=predictions, references=labels)

  return bleu_score, rouge_score, accuracy_score, f1_score
```

### ŪēôžäĶ

```python
trainer = SFTTrainer(
    model=model,
    train_dataset=train_data,
    eval_dataset=test_data,
    peft_config=peft_params,
    args=training_params,
    compute_eval=compute_eval
)
```

```python
trainer.train()

trainer.model.save_pretrained(new_model)

wandb.finish()
```
